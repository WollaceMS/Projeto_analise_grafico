#### Exportações brasileiras para os EUA

```python
df= df.loc[(df['Country'] == 'United States'), :]
display(df)

#variavel alvo = 'US$ FOB'
#variáveis categóricas - City, SH4 Description, SH2 Description, Economic BLock
```

```python
df= df.drop(columns=['SH2 Code', 'SH4 Code'], axis=1)
colunas_numericas = df.select_dtypes(include='number').columns
colunas_categoricas = df.select_dtypes(exclude='number').columns
print(colunas_categoricas)
print(colunas_numericas)
```

```python
df_total = df[['Year', 'US$ FOB']]
df_total = df_total.groupby('Year').sum()
df_total
```

```python
from cycler import cycler


plt.figure(figsize=(14,6))

cores = plt.get_cmap('Set3').colors
cores_ciclo = cycler('color',cores)
plt.rc('axes', prop_cycle=cores_ciclo)
cores_anotacao = cores[4]

plt.plot(df_total.index, df_total['US$ FOB'],  linewidth=2, zorder=0)
plt.xticks(df_total.index)
plt.tick_params(axis='x', length=0)
plt.yticks([])
plt.ylabel('Valor (em bilhões US$)', fontsize=12)
l = list(df_total['US$ FOB'])
for posicao in (-1,0):
    plt.scatter(df_total.index[posicao], l[posicao], zorder=1, color=cores_anotacao)
    plt.annotate(f'{l[posicao]:,}', color=cores_anotacao, xy=(df_total.index[posicao], l[posicao]), xytext=(5,0), textcoords = 'offset points', size=10, weight='bold')


plt.axvspan(df_total.index[6], df_total.index[9], color=cores[6], alpha=0.4, zorder=1)
plt.text(2017, l[0], 'Recuperação', fontsize=11, weight='bold', color=cores_anotacao) # numeros inicias são coordenadas do grafico

plt.title('Exportações para os EUA', fontsize=16, color='gray', loc='left')
plt.box(False)
plt.show()
```

#### Exportações de metais

```python
mosaico='AB;CD' #cada letra é uma posição. ; pula a linha
fig = plt.figure(figsize=(12,10))
axs= fig.subplot_mosaic(mosaico)

novo_df = df[df['SH2 Description'].isin(['Aluminium and articles thereof', 'Iron and steel', 'Miscellaneous articles of base metal', 'Zinc and articles thereof'])]
agrupado = novo_df.groupby(['Year', 'SH2 Description']).sum()[['US$ FOB']].reset_index()
aluminio = agrupado[agrupado['SH2 Description'] == 'Aluminium and articles thereof']
aluminio = aluminio.set_index('Year')['US$ FOB'] # seleciona apenas colunas de ano e US$ FOB

cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)


barras = axs['A'].barh(aluminio.index, aluminio)
axs['A'].set_frame_on(False) #sumir com o quadrado que cerca o gráfico

axs['A'].set_xticks([]) #sumir com os marcadores do eixo x
axs['A'].set_yticks(aluminio.index)
axs['A'].tick_params(axis='y', length=0) #sumir com o tracinho e deixar só o ano

axs['A'].bar_label(barras, label=aluminio ,padding = -24, fontsize=10, fontweight='bold', color='white') # numeros em cima das barras e alterar a fonte
#padding - espaçamento entre o número e a barra.


#mudando a cor da maior barra
barras[1].set_color(c='lightblue')
    
axs['A'].set_title('Valor de alumínio exportado em bilhões de US$')


#grafico B
ferro = agrupado[agrupado['SH2 Description'] == 'Iron and steel']
ferro = ferro.set_index('Year')['US$ FOB'] 

barras2 = axs['B'].barh(ferro.index, ferro, color=cores[1])
axs['B'].set_frame_on(False) #sumir com o quadrado que cerca o gráfico

axs['B'].set_xticks([]) #sumir com os marcadores do eixo x
axs['B'].set_yticks(ferro.index)
axs['B'].tick_params(axis='y', length=0) #sumir com o tracinho e deixar só o ano

axs['B'].bar_label(barras2, label=ferro ,padding = -24, fontsize=10, fontweight='bold', color='white') # numeros em cima das barras e alterar a fonte
#padding - espaçamento entre o número e a barra.

barras2[8].set_color(c='lightblue')

axs['B'].set_title('Valor de ferro exportado em bilhões de US$')

#grafico C

metal = agrupado[agrupado['SH2 Description'] == 'Miscellaneous articles of base metal']
metal = metal.set_index('Year')['US$ FOB'] 

barras3 = axs['C'].barh(metal.index, metal, color=cores[3])
axs['C'].set_frame_on(False) #sumir com o quadrado que cerca o gráfico

axs['C'].set_xticks([]) #sumir com os marcadores do eixo x
axs['C'].set_yticks(metal.index)
axs['C'].tick_params(axis='y', length=0) #sumir com o tracinho e deixar só o ano

axs['C'].bar_label(barras3, label=metal ,padding = -24, fontsize=10, fontweight='bold', color='white') # numeros em cima das barras e alterar a fonte
#padding - espaçamento entre o número e a barra.

barras3[5].set_color(c='lightblue')

axs['C'].set_title('Valor de artigos de metal exportado em milhões de US$')

#grafico D
def formatar(valor):
    valor = f'US${valor:,}'
    return valor

zinco = agrupado[agrupado['SH2 Description'] == 'Zinc and articles thereof']
zinco = zinco.set_index('Year')['US$ FOB'].sort_values()

axs['D'].scatter(zinco.index, zinco.apply(formatar),c=zinco, cmap='Reds',s=72)
axs['D'].set_frame_on(False)#sumir com o quadrado que cerca o gráfico

axs['D'].tick_params(axis='x', length=0) #sumir com o tracinho e deixar só o ano 
axs['D'].set_title('Valor de zinco exportado em milhões de US$')

plt.subplots_adjust(wspace=0.5, hspace=0.3, top=0.9)
fig.suptitle('Exportações de materiais metáticos', color='gray', fontsize=16)
plt.show()
```

#### Quais os produtos mais exportados ao longo de todo o período?

Queremos saber os produtos com maior valor US$ de exportação

```python
produtos = df.groupby('SH2 Description').sum(numeric_only=True)
produtos_mais_exportados = produtos[['US$ FOB']]
ordena_por_valor = produtos_mais_exportados.sort_values(by='US$ FOB', ascending=False)
top5 = ordena_por_valor[:5]
plt.style.use('fast')
lista_valores = []
for numero in (top5.values):
    lista_valores.append(numero[0])
destaque = [0.1, 0, 0, 0, 0]
plt.pie(lista_valores, labels = top5.index,explode=destaque, autopct='%.1f',shadow=True)
plt.title('TOP 5 produtos mais exportados', fontsize=16, color='gray')
plt.show()
```

### Qual cidade mais exportou para os EUA em cada ano?
Em valores US$

```python
agrup = df.groupby(['Year', 'City']).sum()[['US$ FOB']].reset_index()

sns.set_theme(style='dark')
id_maiores_valores_por_ano = agrup.groupby('Year')['US$ FOB'].idxmax()

# filtrando
top_cities = agrup.loc[id_maiores_valores_por_ano]
#display(top_cities)
fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(top_cities, x='Year', y='US$ FOB', hue='City', width=1, ax=ax) #width - largura da barra

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.) #borderaxespad - O espaço entre os eixos e a borda da legenda, em unidades de tamanho de fonte.
plt.title('Cidades com maior exportação em US$', fontsize=16, color='gray', loc='center')
plt.show()
```

#### SP vs RJ

```python
sp = df[df['City'].str.contains('SP')]

agrupamento_sp = sp.groupby('City').sum()[['US$ FOB']].reset_index()
agrupamento_sp = agrupamento_sp.sort_values(by='US$ FOB', ascending=False)
top10_sp = agrupamento_sp[:10]
```

```python
fig, ax  = plt.subplots(figsize=(14,8))
sns.set_theme(palette='tab20', style='dark')
h1= sns.barplot(top10_sp, x='City', y='US$ FOB', width=0.8, ax = ax)
plt.xticks(rotation=45)
plt.xlabel('')
plt.yticks([])
plt.ylabel('')
for barra in h1.containers:
    h1.bar_label(barra, label_type='edge', labels= [f'US${b.get_height():,.0f}' for b in barra], color='black', fontsize=9)
plt.title('Maiores cidades exportadoras de SP de 2010 a 2020', fontsize=16, color='gray', loc='left')
  
plt.show()
```

```python
rj = df[df['City'].str.contains('RJ')]
agrupamento_rj = rj.groupby('City').sum()[['US$ FOB']].reset_index()
agrupamento_rj = agrupamento_rj.sort_values(by='US$ FOB', ascending=False)
top10_rj = agrupamento_rj[:10]
top10_rj
```

```python
fig, ax  = plt.subplots(figsize=(14,8))
plt.style.use('bmh')
h2= sns.barplot(top10_rj, x='City', y='US$ FOB', width=0.8, ax = ax)
plt.xticks(rotation=45)
plt.xlabel('')
plt.yticks([])
plt.ylabel('')
for barras in h2.containers:
    h2.bar_label(barras, label_type='edge', labels= [f'US${barra.get_height():,.0f}' for barra in barras], color='black', fontsize=9)
plt.title('Maiores cidades exportadoras do RJ de 2010 a 2020', fontsize=16, color='gray', loc='left')
plt.show()
```

#### Em cada ano, quais cidades mais exportaram mais em termos de peso líquido (Net weight)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular pesos
peso = df.groupby(['Year', 'City']).sum()['Net Weight'].reset_index()
top_exportadores = peso.loc[peso.groupby('Year').idxmax()['Net Weight'], :]

# Definir o tamanho do gráfico
plt.figure(figsize=(12, 6))

# Criar o gráfico de dispersão
sns.scatterplot(data=top_exportadores, x='Year', y='Net Weight', hue='City', alpha=0.7, s=100)

# Adicionar título e rótulos
plt.title('Top Exportadores por Ano', fontsize=16)
plt.xlabel('Ano', fontsize=14)
plt.ylabel('Peso Líquido (Net Weight)', fontsize=14)

# Ajustar marcas do eixo x
plt.xticks(top_exportadores['Year'].unique(), rotation=45)

# Melhorar a legenda
plt.legend(title='Cidade', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

#### Produtos agrícolas

```python
agricolas = df[df['SH2 Description'].isin(['Cereals',
 'Cocoa and cocoa preparations',
 'Coffee, tea, maté and spices',
 'Edible fruit and nuts; peel of citrus fruits or melons',
 'Oil seeds and oleaginous fruits; Grains, Seeds, others',])]
agricolas_ano = agricolas.groupby(['Year', 'SH2 Description']).sum()[['US$ FOB']].reset_index()

plt.figure(figsize=(16,7))
ord_prod =agricolas_ano.sort_values(by='SH2 Description')
sns.barplot(data=ord_prod, x='Year', y='US$ FOB', hue='SH2 Description', alpha=0.7, width=1)
plt.legend(title='Produtos agrícolas', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title('Exportações de produtos agrícolas', color='gray');
```
